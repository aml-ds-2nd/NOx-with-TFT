```python
cor = z.corr()
```

```python
c = cor['STACK NOX'].reset_index()
c.columns=['a','b']
```

```python
c.sort_values(by='b', ascending=False, inplace=True)
```

```python
cx = c['a'].to_list()
cy = c['b'].to_list()
```

```python
x = np.array(cx)
y = np.array(cy)
```

```python
plt.figure(figsize=(15,8))
threshold = 0

above_threshold = y > threshold
plt.bar(x[above_threshold], y[above_threshold], color='blue')
below_threshold = np.logical_not(above_threshold)
plt.bar(x[below_threshold], y[below_threshold], color='orange') 

plt.xticks(rotation=90)
plt.axhline(y=0.5, xmin=0, xmax=1, color='red', linestyle='--', alpha=0.8)
plt.axhline(y=-0.5, xmin=0, xmax=1, color='red', linestyle='--', alpha=0.8)
plt.tight_layout()
plt.show()
```
